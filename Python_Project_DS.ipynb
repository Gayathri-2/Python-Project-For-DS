**Question 1 - Extracting Tesla Stock Data Using yfinance **

```python
import yfinance as yf

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max") # You can adjust the period as needed
tesla_data.reset_index(inplace=True)
tesla_data.head()  # Show the first few rows to check
```

Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
import requests
from bs4 import BeautifulSoup

# URL for Tesla's financials (example)
url = "https://www.marketwatch.com/investing/stock/tsla/financials"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the revenue table and extract data
# (You'll need to inspect the page and find the appropriate tags/classes)
revenue_table = soup.find('table', {'class': 'table'})
print(revenue_table)  # You may want to parse the table further
```
